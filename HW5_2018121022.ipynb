```python
#3번
# 평균 문제
print("""문제: 수강생 1 의 시험성적을 list 로 선언하고 평균 점수를 구한 뒤,
80점이 넘으면 합격, 80점보다 낮으면 불합격이 출력되게 하시오.""")

A = 92 # A는 A과목 시험 성적
B = 96 # B는 B과목 시험 성적
C = 100 # C는 C과목 시험 성적

score_list = [A, B, C]
print(f'시험성적 리스트 = {score_list}')

score_average = sum(score_list)/len(score_list)
print(f'평균 = {score_average}')

if score_average > 80:
    print('합격')
else:
    print('불합격')
```

```python
#6번
print('''문제) for문을 사용하여 1부터 100까지 숫자를 출력하고,
역순으로도 출력해보자''')

for i in range(1, 101):
    print(i)

# 역순
for i in range(100, 0, -1):
    print(i)

print('''문제) for문을 사용하여 1부터 100까지 숫자를 출력하고,
숫자를 더한 값을 출력하라.''')

# 결과 변수 선언
result = 0

for i in range(1, 101):
    result += i

print(result)
```

```python
#9번
print('''문제) 30보다 작거나 같은 짝수끼리의 곱과 홀수끼리의 곱을 구하고 대소를 비교해라.''')

even_mul = 1
odd_mul = 1

for i in range(1, 31):
    if i % 2 == 0:
        even_mul *= i
    else:
        odd_mul *= i

if even_mul > odd_mul:
    result = f"짝수곱 {even_mul}이 홀수곱 {odd_mul}보다 크다"
elif even_mul == odd_mul:
    result = f"홀수곱 {odd_mul}이 짝수곱 {even_mul}보다 크다"
else:
    result = f"짝수곱 {even_mul}과 홀수곱 {odd_mul}의 크기는 같다"
print(result)
```

```python
#10번
print('''문제) 5 명까지 탑승할 수 있는 놀이기구의 키 제한이 150cm 이상, 200cm 이하일 때 주어진 목록을 이용해 탑승 검사 코드를 만들어보자.''')

max_number = 5
min_height = 150
max_height = 200
user_height_dict = {'1 번': 139, '2 번': 170, '3 번': 178, '4 번': 205, '5 번': 168, '6 번': 149, '7 번': 150, '8 번': 166, '9 번': 180, '10 번': 200}

for k in user_height_dict:
    v = user_height_dict[k]

r = {}
for k in user_height_dict:
    if len(r) == max_number:
        print(f"{max_number} 명 모두 찼습니다. 다음 차례를 이용해 주세요.")
    else:
        v = user_height_dict[k]
        if min_height <= v <= max_height:
            r[k] = v
            print(f"{k}분은 탑승하셔도 됩니다. 현재: {len(r)}명 탑승")
        else:
            if v < min_height:
                print(f"{k}분은 키가 작아 탑승하실 수 없습니다.")
            elif v > max_height:
                print(f"{k}분은 키가 커 탑승하실 수 없습니다.")
```

```python
#12번
print('''문제) while문을 사용하여
1부터 1000까지의 자연수 중 3의 배수의 합을 구해 보자''')

# 첫 번째 방법

i = 1
s = 0

while True:
    if i > 1000:
        break
    
    if i % 3 == 0:
        s += i
    
    i += 1

print(s)

# 두 번째 방법

i = 1
s = 0

while i <= 1000:
    if i % 3 == 0:
        s += i
    i += 1

print(s)
```
